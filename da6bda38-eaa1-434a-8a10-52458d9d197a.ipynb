### Шаг 2.1 Заполнение пропусков

Заполним пропуски NaN в столбцах 'total_income'. Можно найти медиану по все столбцу, однако это будет не корректно, т.к. кредит берут разные группы по типам занятости. Поэтому используем группировку по типам занятости ('income_type') и найдем медианное значение для каждой группы по ежемесячному доходу ('total_income'). Таким образом, мы выведем медиану для каждой группы клиентов, что будет более корректно.

```python
df.groupby('income_type')['total_income'].transform('median').head(15) #группировка по 'income_type', медиана по 'total_income'
```

Заполняем столбец 'total_income'. Проверяем таблицу.

```python
df['total_income'] = df['total_income'].fillna(df.groupby('income_type')['total_income'].transform('median'))
df.head(15)
```

Пропуски в столбце 'total_income' заполнены медианными значениями в соответствии с типом занятости клиента.

### Шаг 2.2 Проверка данных на аномалии и исправления.

В данных столбца 'children' есть значения от -1 до 20. Определим варианты заполнения столбца:

```python
df.groupby('children')['children'].count() # используем группировку по детям и подсчет для каждого варианта
```

Итак: в таблице есть 47 строк, в которых указан -1 ребенок(не допустимое значение) и 76 строк со значение 20 детей (маловероятно). Варианты с количество детей от 3 до 5 вполне допустимы (многодетные семьи). Причина ошибки может быть как в человеке, так и в сбое при сборе данных. В любом случае, процент ошибки мал, что не должно сильно повлиять на результат дальнейшей обработки. Однако ее необходимо исправить: значение строк с -1 заменим на 0 детей, а строки со значением 20 заменим на 2. 

```python
df.loc[df['children'] == -1, ['children']] = 0 # замена для строк со значением -1 на 0 детей
df.loc[df['children'] == 20, ['children']] = 2 # замена для строк со значением 20 на 2 ребенка
df.groupby('children')['children'].count() # проверка 
```

### Шаг 2.5. Категоризация целей кредита.

```python
def purpose_category(purpose):
    """
    Возвращает новую категорию по значению столбца 'purpose', используя правила:
    - 'операции с автомобилем', если в purpose есть подстрока 'автомобил';
    - 'операции с недвижимостью', если в purpose есть подстрока 'жиль' or 'недвижимост';
    - 'проведение свадьбы', если в purpose есть подстрока 'свадьб';
    - 'получение образование', если в purpose есть подстрока 'образован'.
    """
    
    if 'автомобил' in purpose:
        return 'операции с автомобилем'
    if 'жиль' in purpose or 'недвижимост' in purpose:
        return 'операции с недвижимостью'
    if 'свадьб' in purpose:
        return 'проведение свадьбы'
    if 'образован' in purpose:
        return 'получение образование'
    return 'уточните чель кредита' 

df['purpose_category'] = df['purpose'].apply(purpose_category)

display(df.head(15))
```

```python
df.groupby('purpose_category')['purpose_category'].count()
```

Проверка на явные дубликаты после выявления неявных дубликатов:

```python
print(df.duplicated().sum()) # подсчет явных дубликатов
```

Очистим таблицу от явных дубликатов. Перезапишем индексы строк заново.

```python
df = df.drop_duplicates().reset_index(drop=True) #удаляем явные дубликаты, индексы прописываются заново 
print(df.duplicated().sum()) # проверка 
```

### Шаг 2.7. Категоризация дохода.

Дополним таблицу столбцом 'total_income_category', в котором будут указаны категории кредитополучателей в соотвествии с доходом. Для этого создадим функцию:

```python
def total_income_category(total_income):
    """
    Возвращает новую категорию по значению столбца 'total_income', используя правила:
    - если доход от 0 до 30000, то категория Е;
    - если доход от 30001 до 50000, то категория D;
    - если доход от 50001 до 200000, то категория C;
    - если доход от 200001 до 1000000, то категория B;
    - если доход от 1000001 и выше, то категория А.
    """
    
    if 0 <= total_income <= 30000:
        return 'E'
    if 30001 <= total_income <= 50000:
        return 'D'
    if 50001 <= total_income <= 200000:
        return 'C'
    if 200001 <= total_income <= 1000000:
        return 'B'
    if 1000001 <= total_income:
        return 'A'
    return 'уточните ежемесячный доход' 

df['total_income_category'] = df['total_income'].apply(total_income_category)

display(df.head(15))
```

Вывод по гипотезе: предположение подтвердилось. Кредитополучатели с детьми склонны к просрочке по выплате кредита.
