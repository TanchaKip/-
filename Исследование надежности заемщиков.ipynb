```python
def purpose_category(purpose):
    """
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –Ω–æ–≤—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏—é –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é —Å—Ç–æ–ª–±—Ü–∞ 'purpose', –∏—Å–ø–æ–ª—å–∑—É—è –ø—Ä–∞–≤–∏–ª–∞:
    - '–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–º', –µ—Å–ª–∏ –≤ purpose –µ—Å—Ç—å –ø–æ–¥—Å—Ç—Ä–æ–∫–∞ '–∞–≤—Ç–æ–º–æ–±–∏–ª';
    - '–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å—é', –µ—Å–ª–∏ –≤ purpose –µ—Å—Ç—å –ø–æ–¥—Å—Ç—Ä–æ–∫–∞ '–∂–∏–ª—å' or '–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç';
    - '–ø—Ä–æ–≤–µ–¥–µ–Ω–∏–µ —Å–≤–∞–¥—å–±—ã', –µ—Å–ª–∏ –≤ purpose –µ—Å—Ç—å –ø–æ–¥—Å—Ç—Ä–æ–∫–∞ '—Å–≤–∞–¥—å–±';
    - '–ø–æ–ª—É—á–µ–Ω–∏–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ', –µ—Å–ª–∏ –≤ purpose –µ—Å—Ç—å –ø–æ–¥—Å—Ç—Ä–æ–∫–∞ '–æ–±—Ä–∞–∑–æ–≤–∞–Ω'.
    """
    
    if '–∞–≤—Ç–æ–º–æ–±–∏–ª' in purpose:
        return '–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–º'
    if '–∂–∏–ª—å' in purpose or '–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç' in purpose:
        return '–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å—é'
    if '—Å–≤–∞–¥—å–±' in purpose:
        return '–ø—Ä–æ–≤–µ–¥–µ–Ω–∏–µ —Å–≤–∞–¥—å–±—ã'
    if '–æ–±—Ä–∞–∑–æ–≤–∞–Ω' in purpose:
        return '–ø–æ–ª—É—á–µ–Ω–∏–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ'
    return '—É—Ç–æ—á–Ω–∏—Ç–µ —á–µ–ª—å –∫—Ä–µ–¥–∏—Ç–∞' 

df['purpose_category'] = df['purpose'].apply(purpose_category)

display(df.head(15))
```

–°–æ–∑–¥–∞–¥–∏–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º df_edu —Å –¥–∞–Ω–Ω—ã–º–∏ –∏–∑ —Ç–∞–±–ª–∏—Ü—ã (—Å—Ç–æ–ª–±—Ü—ã 'education' –∏ 'education_id'),  –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥–µ–º —Å–æ–ø–æ—Å—Ç–∞–≤–ª—è—Ç—å –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –∫ –µ–≥–æ —Ç–∏–ø—É (id - '0' —Å–æ–æ—Ç–≤–µ—Ç–≤—É–µ—Ç –≤—ã—Å—à–µ–º—É –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—é –∏ —Ç.–¥.)

### –®–∞–≥ 2.7. –ö–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è –¥–æ—Ö–æ–¥–∞.

–î–æ–ø–æ–ª–Ω–∏–º —Ç–∞–±–ª–∏—Ü—É —Å—Ç–æ–ª–±—Ü–æ–º 'total_income_category', –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥—É—Ç —É–∫–∞–∑–∞–Ω—ã –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∫—Ä–µ–¥–∏—Ç–æ–ø–æ–ª—É—á–∞—Ç–µ–ª–µ–π –≤ —Å–æ–æ—Ç–≤–µ—Å—Ç–≤–∏–∏ —Å –¥–æ—Ö–æ–¥–æ–º. –î–ª—è —ç—Ç–æ–≥–æ —Å–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é:

```python
def total_income_category(total_income):
    """
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –Ω–æ–≤—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏—é –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é —Å—Ç–æ–ª–±—Ü–∞ 'total_income', –∏—Å–ø–æ–ª—å–∑—É—è –ø—Ä–∞–≤–∏–ª–∞:
    - –µ—Å–ª–∏ –¥–æ—Ö–æ–¥ –æ—Ç 0 –¥–æ 30000, —Ç–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è –ï;
    - –µ—Å–ª–∏ –¥–æ—Ö–æ–¥ –æ—Ç 30001 –¥–æ 50000, —Ç–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è D;
    - –µ—Å–ª–∏ –¥–æ—Ö–æ–¥ –æ—Ç 50001 –¥–æ 200000, —Ç–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è C;
    - –µ—Å–ª–∏ –¥–æ—Ö–æ–¥ –æ—Ç 200001 –¥–æ 1000000, —Ç–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è B;
    - –µ—Å–ª–∏ –¥–æ—Ö–æ–¥ –æ—Ç 1000001 –∏ –≤—ã—à–µ, —Ç–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è –ê.
    """
    
    if 0 <= total_income <= 30000:
        return 'E'
    if 30001 <= total_income <= 50000:
        return 'D'
    if 50001 <= total_income <= 200000:
        return 'C'
    if 200001 <= total_income <= 1000000:
        return 'B'
    if 1000001 <= total_income:
        return 'A'
    return '—É—Ç–æ—á–Ω–∏—Ç–µ –µ–∂–µ–º–µ—Å—è—á–Ω—ã–π –¥–æ—Ö–æ–¥' 

df['total_income_category'] = df['total_income'].apply(total_income_category)

display(df.head(15))
```

1. –ß–µ–º –±–æ–ª—å—à–µ –¥–µ—Ç–µ–π —É –∫—Ä–µ–¥–∏—Ç–æ–ø–æ–ª—É—á–∞—Ç–µ–ª—è, —Ç–µ–º —Ä–µ–∂–µ –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç –∫—Ä–µ–¥–∏—Ç –≤ —Å—Ä–æ–∫.

### –ì–∏–ø–æ—Ç–µ–∑–∞ ‚Ññ1

–ü—Ä–æ–≤–µ—Ä–∏–º –ø–µ—Ä–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: "–ß–µ–º –±–æ–ª—å—à–µ –¥–µ—Ç–µ–π —É –∫—Ä–µ–¥–∏—Ç–æ–ø–æ–ª—É—á–∞—Ç–µ–ª—è, —Ç–µ–º —Ä–µ–∂–µ –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç –∫—Ä–µ–¥–∏—Ç –≤ —Å—Ä–æ–∫." 
–°–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É, –≤ –∫–æ—Ç–æ—Ä–æ–π –±—É–¥–µ—Ç –æ—Ç–æ–±—Ä–∞–∂–∞—Ç—å—Å—è –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π ('children') –∏ —Å—É–º–º–∞—Ä–Ω–∞—è –∑–∞–¥–æ–ª–∂–µ–Ω–Ω–æ—Å—Ç—å –ø–æ –ø–æ–≥–∞—â–µ–Ω–∏—é –∫—Ä–µ–¥–∏—Ç–∞ –≤ —Å—Ä–æ–∫ ('debt'). –ù–∞–π–¥–µ–º –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ —Å–æ–æ—Ç–Ω–æ—â–µ–Ω–∏–µ ('ratio %') –∑–∞–¥–æ–ª–∂–µ–Ω–Ω–æ—Å—Ç–∏ –ø–æ –ø–æ–≥–∞—â–µ–Ω–∏—é –∫—Ä–µ–¥–∏—Ç–∞ –≤ —Å—Ä–æ–∫ –∫ –æ–±—â–µ–º—É —á–∏—Å–ª—É –∫—Ä–µ–¥–∏–ø–æ–ª—É—á–∞—Ç–µ–ª–µ–π –≤ –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø–µ ('children').

```python
children_pivot = df.pivot_table(index=['children'], values='debt', aggfunc= ('count', 'sum')) #—Å–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ –ø–æ 'children' –∏ 'debt' 
children_pivot['ratio, %'] = (children_pivot['sum'] / children_pivot['count']) * 100 # —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö
display(children_pivot.head(10))
```
